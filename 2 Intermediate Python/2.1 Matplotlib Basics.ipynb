## Basic Plots with Matplotlib
#### Line Plot
A line plot is commonly used to display data points over a continuous interval or time period. Matplotlibâ€™s plot() function allows you to create line plots with ease.

```python
# Import Matplotlib
import matplotlib.pyplot as plt
# Assign equal number of values for 2 variables
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
# Plot using plt.plot(x-axis, y-axis)
plt.plot(x, y)
# Generate plot visually
plt.show()
```

#### Line Plot
You can add multiple lines to a single plot by calling plot() multiple times with different x and y values.

```python

plt.plot(x, y, label='y = x^2')
plt.plot(x, [xi**1.5 for xi in x], label='y = x^1.5')
plt.legend()
plt.show()
```

You can also customize plots

```python
plt.plot(x, y, color='purple', linestyle='-', linewidth=0.5, marker='o', markersize=5, label='y = x^2')
plt.plot(x, [xi**1.5 for xi in x], color = 'red', linestyle = '--', linewidth = 2,  marker='o', markersize = 5, label='y = x^1.5')
plt.show()
```

Scatterplot

```python
import numpy as np
# Scatterplot
x = np.random.rand(50)
y = np.random.rand(50) # Generate 50 random point for x and y
plt.title("Basic Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis") # Give title and labels
plt.scatter(x, y)
plt.show()
```

```python
# Customizing Scatterplots
# Scatter plot with customized marker size and color
plt.scatter(x, y, s=100, c='red', alpha=0.5)  # Marker size of 100, red color, and 50% transparency

# Add title and labels
plt.title("Customized Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
```

#### Histograms
Histograms show the distribution of data by grouping values into bins, which is especially helpful for understanding frequency and spread.

```python
# Histogram
data = [1, 1, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data)
plt.show()
# Histogram: Bins
plt.hist(data, bins=4)
plt.show()
# Histogram: Compare
data1 = [1, 2, 2, 3, 3, 3]
data2 = [3, 4, 4, 4, 5, 5]
plt.hist(data1, alpha=0.5, label='data1')
plt.hist(data2, alpha=0.5, label='data2')
plt.legend()
plt.show()
```

#### Choose the Right Plot
Understanding which plot to use for different data types and analysis goals is essential.
- Line Plot: Best for showing trends over time.
- Scatter Plot: Best for exploring relationships between two variables.
- Histogram: Best for understanding the distribution of a single variable.

Consider the goal of your analysis:
- Trends: Line plot
- Correlation: Scatterplot
- Distribution: Histogram

#### Customization
To make plots more informative, Matplotlib offers several customization options.

```python
# Data for scatter plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

# Create a figure with the desired size
plt.figure(figsize=(10, 5))

# Adjust ticks
plt.xticks([1, 2, 3, 4, 5])
plt.yticks([10, 20, 30, 40, 50])

# Plotting the scatter plot
plt.scatter(x, y, color='red')

# Adding gridlines
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
data = np.random.normal(0, 1, 1000)

# Line Plot
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='--')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='-.')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot of Sine and Cosine")
plt.legend()
plt.grid(True)
plt.show()

# Scatter Plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y1, color='purple', label='sin(x)')
plt.scatter(x, y2, color='orange', label='cos(x)')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Sine and Cosine")
plt.legend()
plt.show()

# Histogram
plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, color='gray', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Data")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the CSV file with 'date' as the index
sf = pd.read_csv(r"C:\Users\Renz\.cache\kagglehub\datasets\abubakkar123\walmart-stores-weekly-sales-forecasting\versions\1\Walmart.csv")
# Convert the Date column to be recognizeable as date by the dataframe
sf['Date'] = pd.to_datetime(sf['Date'], dayfirst = True)

# Filter the data for Store 1 and for the year 2010 and onwards
store_1_data = sf[(sf['Store'] == 1) & (sf['Date'].dt.year == 2010)]
store_2_data = sf[(sf['Store'] == 2) & (sf['Date'].dt.year == 2010)]
store_3_data = sf[(sf['Store'] == 3) & (sf['Date'].dt.year == 2010)]
store_4_data = sf[(sf['Store'] == 4) & (sf['Date'].dt.year == 2010)]

# Plot the data
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.plot(store_1_data['Date'], store_1_data['Weekly_Sales'], label='Weekly Sales for Walmart Store 1', color = 'Blue')
plt.plot(store_2_data['Date'], store_2_data['Weekly_Sales'], label='Weekly Sales for Walmart Store 2', color = 'Red')
plt.plot(store_3_data['Date'], store_3_data['Weekly_Sales'], label='Weekly Sales for Walmart Store 3', color = 'Green')
plt.plot(store_4_data['Date'], store_4_data['Weekly_Sales'], label='Weekly Sales for Walmart Store 4', color = 'Purple')

# Customize
plt.title('Weekly Sales for Store 1-4 in 2010')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
